```python
import pygame
import random
import os
pygame.init()
pygame.mixer.init()


# Colors
white = (255, 255, 255)
red = (255, 0, 0)
black = (0, 0, 0)
diffcolor = (13,165,222)
colour = (12,124,32)

# Creating window
screen_width = 800
screen_height = 400
gameWindow = pygame.display.set_mode((screen_width, screen_height))
bgimg = pygame.image.load(r"C:\Users\swatantra gupta\PycharmProjects\snakevsfoodgame2\theme.mp3\mainwall.jpg.jpg")
bgimg = pygame.transform.scale(bgimg,(screen_width,screen_height)).convert_alpha()
pygame.display.update()

cgimg = pygame.image.load(r"C:\Users\swatantra gupta\PycharmProjects\snakevsfoodgame2\theme.mp3\ground.jpg.jpg")
cgimg = pygame.transform.scale(cgimg, (screen_width, screen_height)).convert_alpha()
pygame.display.update()

dgimg = pygame.image.load(r"C:\Users\swatantra gupta\Desktop\python\lallu.jpg.jpg")
dgimg = pygame.transform.scale(dgimg, (screen_width, screen_height)).convert_alpha()
pygame.display.update()
        
# Game Title
pygame.display.set_caption("Snake vs food")
pygame.display.update()
clock = pygame.time.Clock()
font = pygame.font.SysFont("segeoprint", 35)
pygame.display.update()
def text_screen(text, color, x, y):
    screen_text = font.render(text, True, color)
    gameWindow.blit(screen_text, [x,y])


def plot_snake(gameWindow, color, snk_list, snake_size):
    for x,y in snk_list:
        pygame.draw.rect(gameWindow, black, [x, y, snake_size, snake_size])

def welcome():
    exit_game = False
    while not exit_game:
        gameWindow.fill((143,222,110))
        gameWindow.blit(dgimg , (0,0))
        text_screen(" SNAKE VS FOOD  ",black,290,90)
        text_screen("press space to play  ",black,290,300) 
     

        for event in pygame.event.get():
            pygame.mixer.music.load(r"C:\Users\swatantra gupta\Desktop\python\jungle.mp3.mp3")
            pygame.mixer.music.play(-1)  
            pygame.mixer.music.set_volume(0.7)

            if event.type == pygame.QUIT:
                exit_game = True

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    pygame.mixer.music.load(r"C:\\Users\\swatantra gupta\\PycharmProjects\\snakevsfoodgame2\\theme.mp3\\comedy.mp3.mp3")
                    pygame.mixer.music.play(-1)
                    pygame.mixer.music.set_volume(0.7)
                    gameloop()



        pygame.display.update()
        clock.tick(60)


# Game Loop
def gameloop():
    # Game specific variables
    exit_game = False
    game_over = False
    snake_x = 45
    snake_y = 55
    velocity_x = 0
    velocity_y = 0
    snk_list = []
    snk_length = 1
    if(not os.path.exists("highscore.txt")):
        with open("highscore.txt","w") as f:
            f.write("0")

    with open("highscore.txt","r") as f:
        highscore = f.read()

    food_x = random.randint(20, screen_width/2)
    food_y = random.randint(20, screen_height/2)
    pygame.display.update()
    score = 0
    init_velocity = 3
    snake_size = 13
    food_size = 13
    fps = 60
    while not exit_game:
        if game_over:
            with open("highscore.txt","w") as f:
                f.write(str(highscore))
            gameWindow.fill(diffcolor)
            gameWindow.blit(cgimg,(0,0))
            text_screen("Game Over! ", black, 320,150)
            text_screen("press enter to play again ! ", black , 240, 200)

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RETURN:
                        welcome()

        else:

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RIGHT:
                        velocity_x = init_velocity
                        velocity_y = 0

                    if event.key == pygame.K_LEFT:
                        velocity_x = - init_velocity
                        velocity_y = 0

                    if event.key == pygame.K_UP:
                        velocity_y = - init_velocity
                        velocity_x = 0

                    if event.key == pygame.K_DOWN:
                        velocity_y = init_velocity
                        velocity_x = 0

                    if event.key == pygame.K_q:
                        score = score + 10

                    if event.key == pygame.K_a:
                        init_velocity = init_velocity + 1

                    if event.key == pygame.K_s:
                        init_velocity = init_velocity - 1

            snake_x = snake_x + velocity_x
            snake_y = snake_y + velocity_y

            if abs(snake_x - food_x)<7 and abs(snake_y - food_y)<7:
                score = score + 10
                food_x = random.randint(50, screen_width/2)
                food_y = random.randint(50, screen_height/2)
                pygame.display.update()
                snk_length +=5
                if (score) > int(highscore):
                    highscore = score

            gameWindow.fill(diffcolor)
            gameWindow.blit(bgimg,(0,0))
            text_screen("SCORE: " + str(score)+"     HIGHSCORE: " + str(highscore),black,245,5)
            pygame.display.update()
            pygame.draw.rect(gameWindow, red, [food_x, food_y, snake_size, snake_size])


            head = []
            head.append(snake_x)
            head.append(snake_y)
            snk_list.append(head)

            if len(snk_list)>snk_length:
                del snk_list[0]

            if head in snk_list[:-1]:
                game_over = True
                pygame.mixer.music.load(r"C:\\Users\\swatantra gupta\\PycharmProjects\\snakevsfoodgame2\\theme.mp3\\gameover.mp3.mp3")
                pygame.mixer.music.play()
                pygame.mixer.music.set_volume(0.7)

            if snake_x<0 or snake_x>screen_width or snake_y<0 or snake_y>screen_height:
                game_over = True
                pygame.mixer.music.load(r"C:\\Users\\swatantra gupta\\PycharmProjects\\snakevsfoodgame2\\theme.mp3\\gameover.mp3.mp3")
                pygame.mixer.music.play()
                pygame.mixer.music.set_volume(0.7)
            plot_snake(gameWindow, black, snk_list, snake_size)
        pygame.display.update()
        clock.tick(fps)

    pygame.quit()
    quit()
welcome()
pygame.display.update()
```
